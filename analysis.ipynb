# Phone sales analysis

First, we will need to load the dataset and import the necessities.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Order_details(masked).csv")
df.head()
```

```python
df['Time']  = pd.to_datetime(df["Transaction Date"])
df["Hour"]  = (df['Time']).dt.hour
df.drop(columns=['Transaction Date'],inplace=True)
df
```

```python
time = df['Hour'].value_counts().index.tolist()[:24]
amount = df["Hour"].value_counts().values.tolist()[:24]
sales = pd.DataFrame(({"Hour": time,"Amount of People" : amount}))
sales = sales.sort_values(by="Hour")
sales
```

```python
plt.figure(figsize=(20,10))
plt.plot(sales.Hour,sales["Amount of People"],'r')
plt.xlabel("Hours")
plt.ylabel("Amount of people")
plt.title("Phone sales for each hour")
plt.grid()
plt.show()
```

As you can see around midnight till 10 in the morning,amount of people who buy are the phone are under 30 and the peak is around 12 hour. It drops back again and peak at around 23 hour.
